## The .isin() Meethod

The isin() method checks if the Dataframe contains the specified value(s). It returns a DataFrame similar to the original DataFrame, but the original values have been replaced with True if the value was one of the specified values, otherwise False .

```python
df = pd.read_csv('/Users/ravidonaj/Downloads/All-CSV-ML-Data-Files-Download-master/employees.csv')

df['Start Date'] = pd.to_datetime(df['Start Date'])
df['Last Login Time'] = pd.to_datetime(df['Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype(bool)
df['Gender'] = df['Gender'].astype('category')
df.head()
```

```python
mask1 = df['Team'] == 'Legal'
mask2 = df['Team'] == 'Sales'
mask3 = df['Team'] == 'Product'

df[mask1 | mask2 | mask3]
```

```python
mask = df['Team'].isin(['Legal','Sales','Product','Marketing'])

# .isin() Method will check the given value and return True if the given values matches else returns False
```

```python
df[mask]
```

### The .duolicated() Method

```python
df = pd.read_csv('/Users/ravidonaj/Downloads/All-CSV-ML-Data-Files-Download-master/employees.csv')

df['Start Date'] = pd.to_datetime(df['Start Date'])
df['Last Login Time'] = pd.to_datetime(df['Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype(bool)
df['Gender'] = df['Gender'].astype('category')
df.sort_values('First Name',inplace=True)
df.head()
```

```python
df['First Name'].duplicated()

# Here first values will be written as False and it mark it as not duplicated and next item will be considered as duplicated values andreturn True
```

```python
df['First Name'].duplicated(keep='last')
```

```python
df[df['First Name'].duplicated(keep=False)].head(10)

```

```python
df[~df['First Name'].duplicated(keep=False)].head(10)
```
