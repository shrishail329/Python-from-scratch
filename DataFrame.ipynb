### Sort a DataFrame with the sort_values() Method

```python
df.Name.sort_values()

# sort the values alphabetically 
df.Salary.sort_values()
```

```python
df.sort_values(by='Name')

# by default order is ascending

df.sort_values(by='Name', inplace=True)

df.sort_values(by='Name', inplace=False)
```

```python
df.sort_values('Age')
```

```python
df.Salary.sort_values(na_position='last')

# na_postion --> at the last and first

df.Salary.sort_values(na_position='first')
```

```python
# sort more than one column name 

df.sort_values(by=['Team','Name'],ascending=True)
df.sort_values(by=['Team','Name'],ascending=False)

df.sort_values(by=['Team','Name'],ascending=[True,False])
# Team is in ascending order and Name column is descening order
```

```python
df.sort_values(['Position','Salary'])

df.sort_values(['Position','Salary'],ascending=[True,False])
```

### Rank values with the rank Method

```python
df = pd.read_csv('/Users/ravidonaj/Downloads/All-CSV-ML-Data-Files-Download-master/nba.csv').dropna(how='all')
df['Salary'] = df['Salary'].fillna(0)
df.head()
```

```python
df.Salary.rank()

# rank bais 
```

```python
df['Salary_rank'] = df.Salary.rank(ascending=False).astype('int')
```

```python
df.head()
```

```python
df.sort_values('Salary',ascending=False)
```
