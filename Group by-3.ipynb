### Group by

The groupby() method allows you to group your data and execute functions on these groups.

Pandas groupby is used for grouping the data according to the categories and apply a function to the categories. It also helps to aggregate data efficiently.

Pandas groupby is used for grouping the data according to the categories and apply a function to the categories. 

It also helps to aggregate data efficiently. 

Pandas dataframe. groupby() function is used to split the data into groups based on some criteria

Groupby mainly refers to a process involving one or more of the following steps they are: 

#### Splitting : It is a process in which we split data into group by applying some conditions on datasets.
#### Applying : It is a process in which we apply a function to each group independently
#### Combining : It is a process in which we combine different datasets after applying groupby and results into a data structure

### Splitting Data into Groups

Splitting is a process in which we split data into a group by applying some conditions on datasets. In order to split the data, we apply certain conditions on datasets. In order to split the data, we use groupby() function this function is used to split the data into groups based on some criteria. Pandas objects can be split on any of their axes. The abstract definition of grouping is to provide a mapping of labels to group names. Pandas datasets can be split into any of their objects. There are multiple ways to split data like: 
 

1. obj.groupby(key)
2. obj.groupby(key, axis=1)
3. obj.groupby([key1, key2])

## The .agg() Method

```python
df = pd.read_csv('/Users/ravidonaj/Downloads/All-CSV-ML-Data-Files-Download-master/fortune1000.csv',index_col='Rank')
sector = df.groupby('Sector')
df.head()
```

```python
sector.mean().head()
```

```python
sector.agg({'Revenue':'sum',
          'Profits':'sum',
          'Employees':'sum'})
```

```python
sector.agg({'Revenue':'sum',
          'Profits':'mean',
          'Employees':'sum'})
```

```python
sector.agg(['size','sum','mean'])

# agg() aggregates the numeric values 
# use list to aggreagte more than one aggregation function 
```

```python
sector.agg({'Revenue':['sum','mean'],
          'Profits':'mean',
          'Employees':'sum'})

# here revenue column is having both sum() and mean()
```

## Iterating through Groups

```python
df = pd.read_csv('/Users/ravidonaj/Downloads/All-CSV-ML-Data-Files-Download-master/fortune1000.csv',index_col='Rank')
sectors = df.groupby('Sector')
df.head()
```

```python
df = pd.DataFrame(columns = df.columns)
```

```python
df
```

```python
for sector, data in sectors:
    highest_revenue_in_group = data.nlargest(1,'Revenue')
    df = df.append(highest_revenue_in_group)
```
