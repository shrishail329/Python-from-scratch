The concat() function is used to concatenate pandas objects along a particular axis with optional set logic along the other axes.

## Inner Joins, Part -1 

Inner Join: Inner join is the most common type of join you’ll be working with. It returns a dataframe with only those rows that have common characteristics. This is similar to the intersection of two sets.

```python
week1 = pd.read_csv('/Users/ravidonaj/Downloads/All-CSV-ML-Data-Files-Download-master/Restaurant - Week 1 Sales.csv')

week2 = pd.read_csv('/Users/ravidonaj/Downloads/All-CSV-ML-Data-Files-Download-master/Restaurant - Week 2 Sales.csv')

customers = pd.read_csv('/Users/ravidonaj/Downloads/All-CSV-ML-Data-Files-Download-master/Restaurant - Customers.csv')

foods = pd.read_csv('/Users/ravidonaj/Downloads/All-CSV-ML-Data-Files-Download-master/Restaurant - Foods.csv')
```

```python
week1.head(3)
```

```python
week2.head(3)
```

```python
week1.merge(week2,how='inner',on='Customer ID',suffixes=[' - week1',' - week2'])

# merge the week1 and week2 on customer ID column 
# it will return the common column from customer ID
# we can chnage the default food_x values 
# Join based on common columns
```

```python
week1[week1['Customer ID']==155]

# Here on week same customer ordered the different food items 
```

```python
week2[week2['Customer ID']==155]
```

## Inner Joins, Part -2 

```python
week1.merge(week2,how='inner',on=['Customer ID','Food ID'])

## Here it return when both the conditions must satisfay
## Means same customers from both the weeks ordred the same food 
```

```python
week1[week1['Customer ID']==578]
```

```python
week2[week2['Customer ID']==578]

# same customer order the same food twice in a week
```

## Outer Joins

It comibines the avlues either or both together

```python
week1.head(3)
```

```python
week2.head(3)
```

```python
week1.merge(week2,how='outer',on='Customer ID')
```

Here we cam observe above, it return the both data values either it matches or not,
If it does not match, it return as NaN

```python
merged = week1.merge(week2,how='outer',on='Customer ID',indicator=True)
merged
# Indicator tells 
```

If True, adds a column to the output DataFrame called "_merge" with
    information on the source of each row

```python
merged['_merge'].value_counts()
#_merge show the from which the values are
```

```python
mask = merged['_merge'].isin(['left_only','right_only'])

## here we get the only from left and right values of the data frame,
# we cut the both values from the dataframe 
```

```python
merged[mask]
```

## Merging by Index with the left_index and right_index Parameter

```python
week1 = pd.read_csv('/Users/ravidonaj/Downloads/All-CSV-ML-Data-Files-Download-master/Restaurant - Week 1 Sales.csv',index_col='Food ID')

week2 = pd.read_csv('/Users/ravidonaj/Downloads/All-CSV-ML-Data-Files-Download-master/Restaurant - Week 2 Sales.csv')

customers = pd.read_csv('/Users/ravidonaj/Downloads/All-CSV-ML-Data-Files-Download-master/Restaurant - Customers.csv',index_col='ID')

foods = pd.read_csv('/Users/ravidonaj/Downloads/All-CSV-ML-Data-Files-Download-master/Restaurant - Foods.csv')
```

```python
week1.head(3)
```

```python
customers
```

we have given index as food id and id for both the dataframe

```python
sales = week1.merge(customers,how='left',left_on='Customer ID',right_index=True)
sales
```

Here we have merged the column with row index by using right_index()

```python
sales.merge(foods, how='left', left_on='Food ID', right_index=True)
```

```python
week1.merge(week2, how='left',left_index=True,right_index=True)
```
