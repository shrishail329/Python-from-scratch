## A Review of the .apply() Method on Single Columns

```python
df = pd.read_csv('/Users/ravidonaj/Downloads/All-CSV-ML-Data-Files-Download-master/jamesbond.csv',index_col='Film')
df.sort_index(inplace=True)
df.head()
```

```python
def convert_to_string_add_miillions(numbers):
    return str(numbers) + ' Millions!'
```

```python
df['Box Office'] = df['Box Office'].apply(convert_to_string_add_miillions)
```

```python
df.head()
```

```python
df.head(3)
```

```python
columns  = ['Box Office','Budget','Bond Actor Salary']
for cul in columns:
    df[cul] = df[cul].apply(convert_to_string_add_miillions)
```

```python
df.head(3)
```

## The .apply() Method with Row values

```python
df = pd.read_csv('/Users/ravidonaj/Downloads/All-CSV-ML-Data-Files-Download-master/jamesbond.csv',index_col='Film')
df.sort_index(inplace=True)
df.head()
```

```python
def good_movies (row):
    
    Actor = row[1]
    Budget = row[4]
    
    if Actor == 'Roger Moore':
        return 'The best'
    elif Actor == 'David Niven' and Budget  >30:
        return 'Enjoyable'
    else:
        return 'I have no idea'
    
    
df.apply(good_movies,axis='columns')
```
